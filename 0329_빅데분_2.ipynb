```python
import pandas as pd
import numpy as np

mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg_a = mpg.query('displ <= 4')
mpg_a.head()
```

```python
mpg_b = mpg.query('displ <= 5')
mpg_b.head()
```

```python
mpg_a['hwy'].mean()
```

```python
mpg_b['hwy'].mean()
```

```python
if mpg_a['hwy'].mean() > mpg_b['hwy'].mean() :
    print('배기량 4 이하 자동차의 고속도로 연비가 높다')
elif mpg_a['hwy'].mean() < mpg_b['hwy'].mean() :
    print('배기량 4 이하 자동차의 고속도로 연비가 높다')
```
