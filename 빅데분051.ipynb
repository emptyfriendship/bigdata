# Lab 5 : 혼자서 해보기(mpg 데이터 분석)

```python
import pandas as pd
mpg = pd.read_csv('mpg.csv')
mpg.head(3)
```

```python
mpg.groupby('category').agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.groupby('category').agg(mean_cty = ('cty', 'mean')).sort_values('mean_cty', ascending = False)
```

```python
mpg.groupby('manufacturer').agg(mean_hwy = ('hwy', 'mean'))
```

```python
mpg.groupby('manufacturer') \
    .agg(mean_hwy = ('hwy', 'mean'))\
    .sort_values('mean_hwy', ascending = False)\
    .head(3)
```

```python
mpg.query('category == "compact"').head()
```

```python
mpg_new = mpg.query('category == "compact"')
```

```python
mpg_new.groupby('manufacturer')\
        .agg(count_compact = ('category', 'count'))
```

```python
mpg_new.groupby('manufacturer') \
        .agg(count_compact = ('category', 'count')) \
        .sort_values('count_compact', ascending = False) \
        .head(3)
```

```python
mpg_new.value_counts('manufacturer')
```

# Lab 6 : 가로로 합치기

```python
test1 = pd.DataFrame( {'id' : [1, 2, 3, 4, 5, 6],
                      'midterm' : [60, 80, 70, 90, 85, 66]
                      })
test1
```

```python
test2 = pd.DataFrame( {'id' : [1, 2, 3, 4, 5, 7],
                      'finalterm' : [70, 83, 65, 95, 80, 77]
                      })
test2
```

```python
total = pd.merge(test1, test2, how = 'left', on = 'id')
total
```

```python
total = pd.merge(test1, test2, how = 'right', on = 'id')
total
```

```python
name = pd.DataFrame({'nclass' : [1, 2, 3, 4, 5],
                    'teacher' : ['kim', 'lee', 'park', 'choi', 'jung']
                    })
name
```

```python
exam = pd.read_csv('exam.csv')
exam.head(3)
```

```python
exam_new = pd.merge(exam, name, how = 'left', on='nclass')
exam_new
```

# Lab 7 : 세로로 합치기 (concat())

```python
group_a = pd.DataFrame({'id' : [1, 2, 3, 4, 5],
                       'test' : [60, 80, 70, 90, 85]
                       })
```

```python
group_b = pd.DataFrame({'id' : [6, 7, 8, 9, 10],
                       'test' : [70, 83, 65, 95, 80]
                       })
```

```python
group_a
```

```python
group_b
```

```python
group_all = pd.concat([group_a, group_b])
group_all
```

```python
group_all = pd.concat([group_a, group_b], ignore_index=True)
group_all
```

# Lab 9 : 분석 도전

```python
midwest = pd.read_csv('midwest.csv')
midwest.head(3)
```

```python
midwest['ratio'] = (midwest['poptotal'] - midwest['popadults']) / midwest['poptotal'] *100
midwest.head(3)
```

```python
midwest[['county', 'state', 'ratio']].head(3)
```

```python
midwest.sort_values('ratio', ascending=False).head()[['state', 'county','ratio']]
```

```python
import numpy as np
```

```python
midwest['grade'] = np.where(midwest['ratio'] >= 40, 'large',
                   np.where(midwest['ratio'] >= 30, 'middle', 'small'))
midwest.head()[['state', 'county','ratio']]
```

```python
midwest.groupby('grade').agg(count_grade = ('grade', 'count'))
```

```python
midwest.assign(ratio_asgin = midwest['popasian']/midwest['poptotal'] * 100)
```

```python
midwest.assign(ratio_asian = midwest['popasian']/midwest['poptotal'] * 100)\
        .sort_values('ratio_asian')\
        .head(10)[['state', 'county', 'ratio_asian']]
```
