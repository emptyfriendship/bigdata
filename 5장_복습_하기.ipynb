```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)

mpg = pd.read_csv('mpg.csv')
```

```python
mpg_new = mpg.copy()

mpg_new = mpg_new.rename(columns={ 'hwy' : 'highway'})
mpg_new.head()
```

```python
mpg_new = mpg_new.rename(columns={ 'cty' : 'city'})
mpg_new.head()
```

```python
mpg['total'] = (mpg['cty'] + mpg['hwy'])
mpg.head()
```

```python
mpg['mean'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head()
```

```python
mpg.describe()
```

```python
mpg['grade'] = np.where(mpg['total'] >= 47, 'A',
               np.where(mpg['total'] >= 31, 'B', 'C'))
mpg.head()
```

```python
count_grade = mpg['grade'].value_counts().sort_index()
count_grade
```

```python
count_grade.plot.bar(rot=0)
```
